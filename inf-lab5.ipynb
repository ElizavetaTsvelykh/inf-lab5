```python
import math
k = 5  
candidates = [8, 10, 11]
def combinations(k, n):
    return math.factorial(n) // (math.factorial(k) * math.factorial(n - k))
for n in candidates:
    result = combinations(k, n)
    print(f"Количество способов отобрать {k} человек из {n} кандидатов: {result}")
```

```python
import numpy as np
matrices = [
    np.random.randint(1, 10, size=(3, 3)),
    np.random.randint(1, 10, size=(4, 4)),
    np.random.randint(1, 10, size=(5, 5)),
]
def average_excluding_min_max(matrix):
    flat = matrix.flatten()
    min_element = flat.min()
    max_element = flat.max()
    filtered = [x for x in flat if x != min_element and x != max_element]
    if filtered:
        return sum(filtered) / len(filtered)
    else:
        return None 
def check_sequence(arr):
    if all(arr[i] <= arr[i + 1] for i in range(len(arr) - 1)):
        return "возрастающая"
    elif all(arr[i] >= arr[i + 1] for i in range(len(arr) - 1)):
        return "убывающая"
    else:
        return "не монотонная"
averages = [average_excluding_min_max(matrix) for matrix in matrices]
print("Исходные матрицы:")
for i, matrix in enumerate(matrices):
    print(f"Матрица {i + 1}:\n{matrix}")
print("\nСредние значения:", averages)

sequence_type = check_sequence(averages)
print(f"\nПоследовательность средних значений: {sequence_type}")
```
